### TPC4: Aplicação para manipulação de listas de inteiros
- Crie uma aplicação em Python que coloca no monitor o seguinte menu:
    * (1) Criar Lista 
    * (2) Ler Lista
    * (3) Soma
    * (4) Média
    * (5) Maior
    * (6) Menor
    * (7) estaOrdenada por ordem crescente
    * (8) estaOrdenada por ordem decrescente
    * (9) Procura um elemento
    * (0) Sair
- O utilizador irá escolher uma das opções introduzindo o número correspondente;
- Se a opção não for sair, a aplicação executa a operação pretendida, apresenta o resultado e a seguir apresenta de novo o menu;
- Se a opção for sair, a aplicação termina colocando uma mensagem no monitor.

* No desenvolvimento da aplicação deverá ter em atenção o seguinte:
    - A aplicação terá uma variável interna para guardar uma lista de números;
    - Na opção 1, deverá ser criada uma lista de números aleatórios entre 1 e 100 que será guardada na variável interna;
    - Na opção 2, deverá ser criada uma lista com números introduzidos pelo utilizador, que será guardada na variável interna;
    - Nestas primeiras opções, se a variável interna já tiver uma lista, esta será sobreposta/apagada pela nova lista;
    - Na opção 3, será calculada a soma dos elementos na lista no momento;
    - Na opção 4, será calculada a média dos elementos na lista no momento;
    - Na opção 5, será calculado o maior elemento da lista no momento;
    - Na opção 6, será calculado o menor elemento da lista no momento;
    - Na opção 7, a aplicação deverá indicar (Sim/Não) se a lista está ordenada por ordem crescente;
    - Na opção 8, a aplicação deverá indicar (Sim/Não) se a lista está ordenada por ordem decrescente;
    - Na opção 9, a aplicação irá procurar um elemento na lista, se o encontrar deverá devolver a sua posição, devolverá -1 se o elemento não estiver na lista;
    - Se o utilizador selecionar a opção 0, a aplicação deverá terminar mostrando a lista que está nesse momento guardada.

```python
import random
lista_interna = []

# menu
def menu ():
    print("Escolha uma das opções:")
    print("(1) Criar Lista ")  
    print("(2) Ler Lista")  
    print("(3) Soma")
    print("(4) Média")
    print("(5) Maior")
    print("(6) Menor")
    print("(7) Esta ordenada por ordem crescente")
    print("(8) Esta ordenada por ordem decrescente")
    print("(9) Procura um elemento")
    print("(0) Sair")



#opção (1) Criar Lista
def criaLista(N):
    global lista_interna
    lista_interna = [random.randrange(0,101) for x in range(N)]
    


#opção (2) Ler Lista
def leLista(N):
    global lista_interna
    lista_interna = []
    for x in range(N):
        lista_interna.append(int(input("introduza um numero inteiro")))

    
#opção (3) Soma
def somaLista(lista):
   res =  0
   for x in lista:
    res = res + x
   return res
   

#opção (4) Média
def mediaLista(lista):
    soma = 0
    for x in lista:
        soma = soma + x
   
    return soma / len(lista) 


#opção (5) Maior
def maiorLista(lista):
    res = lista [0]
    for x in lista:
        if x > res:
            res = x
    
    return res


#opção (6) Menor
def menorLista(lista):
    res = lista[0]#a lista nao pode ser vazia
    for i in lista:
        if i < res: 
            res = i
    
    return res


#opção (7) Esta ordenada por ordem crescente
def estaOrdenadaCrescente(lista):
    for x in range(len(lista) - 1):
        if lista[x] > lista[x + 1]:
            return False

    return True



#opção (8) Esta ordenada por ordem decrescente
def estaOrdenadaDecrescente(lista):
    for x in range(len(lista) - 1):
        if lista[x] < lista[x + 1]:
            return False

    return True



#opcao (9) Procura um elemento
def procura_um_elemento (lista):
    elemento = input("introduz o elemento que pretendes encontrar")
    a = 0
    while a < len(lista):
        if lista[a] == elemento:
            return a
        a = a + 1
    return -1

#mostrar lista atual
def lista_atual():
    print("lista atual: ", lista_interna)

#funcao principal
def principal ():
    global lista_interna

    while True:
        menu()
        opção = int(input("Que opção queres experimentar?"))

        if opção == 1:
            criaLista(int(input("Introduz o tamanho da lista que quer criar: ")))
        elif opção == 2:
            leLista(int(input("Introduz o tamanho da lista que quer ler: ")))
        elif opção== 3:
            print("Soma:", somaLista(lista_interna))
        elif opção == 4:
            print("Média:", mediaLista(lista_interna))
        elif opção == 5:
            print("Maior elemento:", maiorLista(lista_interna))
        elif opção == 6:
            print("Menor elemento:", menorLista(lista_interna))
        elif opção == 7:
            print("A lista está ordenada por ordem crescente:", estaOrdenadaCrescente(lista_interna))
        elif opção == 8:
            print("A lista está ordenada por ordem decrescente:", estaOrdenadaDecrescente(lista_interna))
        elif opção == 9:
            resultado = procura_um_elemento(lista_interna)
            if resultado == -1:
                print("O elemento não está na lista.")
            else:
                print("O elemento está na posição:", resultado)
        elif opção == 0:
            lista_atual()
            print("A aplicação será terminada.")
            break
        else:
            print("Opção inválida. Por favor, escolha uma opção válida.")

if __name__ == "__main__":
    principal()
```
